```python
x=dataset.iloc[:,0]
y=dataset.iloc[:,1]
x = x.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)

print(x, y)
```

```python
#apply Linear Regression

from sklearn.linear_model import LinearRegression
lregressor = LinearRegression()
lregressor.fit(x,y)
```

```python
ypredict=lregressor.predict(x)
ypredict
```

```python
y
```

```python
plt.scatter(x,y)
plt.plot(x,ypredict,color="green")
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
poly_reg.fit(x_poly,y)
polynomreg=LinearRegression()
polynomreg.fit(x_poly,y)
polypred=polynomreg.predict(x_poly)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,color="pink")
plt.plot(x,polypred,color="aqua")
```

```python
from sklearn.metrics import r2_score
```

```python
lraccuracy = r2_score(y,ypredict)
```

```python
lraccuracy
```

```python
polyaccuracy = r2_score(y,polypred)
```

```python
polyaccuracy
```
